# EU, n = 100, monte carlo

```python
ans = 7.2369476534888735
counter = 0
for i in range(50):
    test1 = HW4.lookBackMCsim(St, r, q, sigma, t, T, 50, 100, 10000, 20)
    test1.sim()
    if ans < test1.interval[2] and ans > test1.interval[0]:
        counter += 1
```

# convergence plot

```python
from matplotlib import pyplot as plt
```

```python
x = list(range(10, 501, 10))
y = []
tLst = []
```

```python
T0 = datetime.now()
for n in x:
    t0 = datetime.now()
    plot_ = HW4.biTree(St, r, q, sigma, t, T, 50, n)
    plot_.compute(bonus1 = True)
    t1 = datetime.now()
    a = t1 - t0
    print(a, end = '\r')
    tLst.append(a.total_seconds())
    y.append(plot_.tree[0][0].payoffDict[50])
T1 = datetime.now()
```

```python
print(T1 - T0)
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, tLst)
```
