# inverse charloskey

```python
from scipy.linalg import cholesky, inv
```

```python
Clambda = np.corrcoef(test.simulator.result_origin[0], rowvar = False)
Clambda
```

```python
Clambda_utri = cholesky(Clambda, lower=False)
Clambda_utri
```

```python
Clambda_utri_inv = inv(Clambda_utri)
Clambda_utri_inv
```

```python
Zilambda = test.simulator.result_origin[0].dot(Clambda_utri_inv)
```

```python
np.corrcoef(Zilambda, rowvar = False)
```
