```python
import biTree
```

```python
S0 = 50
K = 50
r = 0.1
q = 0.05
sigma = 0.4
t = 0.5
```

# black-scholes formula

```python
# S0, r, q, sigma, T, K
bs = biTree.vanillaEuro_bs(S0, r, q, sigma, t, K)
bs.callPrice, bs.putPrice
```

# Monte Carlo Simulation

```python
mcSimCall = biTree.vanillaEuroCall_mcSim(S0, r, q, sigma, t, K, 10000, 20)
mcSimCall.trialstat()
print(mcSimCall.trustinterval[0], mcSimCall.price, mcSimCall.trustinterval[1])
```

```python
mcSimCall.payoffResult.mean(axis = 1)
```

```python
mcSimPut = biTree.vanillaEuroPut_mcSim(S0, r, q, sigma, t, K, 10000, 10)
mcSimPut.trialstat()
print(mcSimPut.trustinterval[0], mcSimPut.price, mcSimPut.trustinterval[1])
```

# convergence plot

```python
priceLst = []
for i in range(1, 500):
    temp = biTree.combiEuro_call(S0, r, q, sigma, t, K, i)
    temp.compute()
    priceLst.append(temp.price)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(20, 10))
ax.plot(priceLst, color='dodgerblue')
ax.axhline(y=bs.callPrice, color='firebrick')
```
